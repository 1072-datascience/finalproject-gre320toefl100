```python
import urllib.request
from bs4 import BeautifulSoup
import csv
import sys
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
```

爬下球員基本資料

```python
#note that tear in url is year (x-1) ~ x 
#stats
for year in range(2011,2019):
    request = urllib.request.Request("https://www.basketball-reference.com/leagues/NBA_"+str(year)+"_totals.html")
    response = urllib.request.urlopen(request)
    html = response.read()
    
    soup = BeautifulSoup(html, 'html.parser')
    attr_list = soup.find_all('table',attrs={'class':'stats_table'})[0].find_all('tr')[0].find_all('th')
    attr_arr = []
    for attr in attr_list:
        attr_arr.append(attr.text)
    print (attr_arr)
    first = True
    count = 0
    player_list = soup.find_all('table',attrs={'class':'stats_table'})[0].find_all('tr')
    with open('./crawler_'+str(int(year)-1)+'-'+str(year)+'_stats.csv','w') as w:
        wr = csv.writer(w,delimiter=',')
        wr.writerow(attr_arr)
        for player in player_list:
            try:
                print (player.find_all('td')[0].text)
            except:
                continue
            stats_arr = []
            for player_arr in player:
                stats_arr.append(player_arr.text)
            wr.writerow(stats_arr)

```

爬下球員薪資資料

```python
#salary
mp = [14, 10, 11, 11, 12, 12, 10] #maxpage
for year in range(2012,2020): #2013s
    with open('./crawler_'+str(year-1)+'-'+str(year)+'_salaries.csv','w') as w:
        ws = csv.writer(w,delimiter=',')
        ws.writerow(['RK','NAME','POS','TEAM','SALARY'])
        for p in range(0, mp[year-2013]): #nowpage
            request = urllib.request.Request("http://www.espn.com/nba/salaries/_/year/"+str(year)+"/page/"+str(p+1))
            response = urllib.request.urlopen(request)
            html = response.read()
            soup = BeautifulSoup(html, 'html.parser')
            row_list = soup.find_all('table',attrs={'class','tablehead'})[0].find_all('tr')
            row_arr = []
            for row in row_list:
                if str(row.attrs['class'][0]) != 'colhead':
                    print (row.find_all('td')[1].text)
                    d = row.find_all('td')
                    ws.writerow([d[0].text,d[1].text.split(', ')[0],d[1].text.split(', ')[1],d[2].text,d[3].text])
            print ('=== PAGE: '+str(p+1)+' FINISH ===')

```

爬下球員進階資料

```python
#adv
for year in range(2011,2019):  #2012-2018
    request = urllib.request.Request("https://www.basketball-reference.com/leagues/NBA_"+str(year)+"_advanced.html")
    response = urllib.request.urlopen(request)
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')
    attr_list = soup.find_all('table',attrs={'class':'stats_table'})[0].find_all('tr')[0].find_all('th')
    attr_arr = []
    for attr in attr_list:
        if attr.text != u'\xa0':
            attr_arr.append(attr.text)
    print (attr_arr)
    first = True
    count = 0
    player_list = soup.find_all('table',attrs={'class':'stats_table'})[0].find_all('tr')
    with open('./crawler_'+str(year-1)+'-'+str(year)+'_stats_advanced.csv','w') as w:
        wr = csv.writer(w,delimiter=',')
        wr.writerow(attr_arr)
        for player in player_list:
            try:
                print (player.find_all('td')[0].text)
            except:
                continue
            stats_arr = []
            for player_arr in player:
                if player_arr.text != '':
                    stats_arr.append(player_arr.text)
            wr.writerow(stats_arr)
            

```

設定球員ID

```python
#pid
from collections import defaultdict
pSet = defaultdict(set)
pid = 0
for y in range(2010,2018):    
    with open('./combine_crawler_'+str(y)+'-'+str(y+1)+'.csv', 'r') as f:
        with open('./NF_'+str(y)+'-'+str(y+1)+'_pid.csv', 'w') as f1:
            ws = csv.writer(f1)
            rs = csv.reader(f)
            first = True
            for r in rs:
                if first:
                    first = False
                    ws.writerow(r);
                elif r[1] not in pSet.keys():
                    pSet[r[1]].add(pid)
                    tmp = r
                    tmp[0] = pid
                    ws.writerow(tmp)
                    pid += 1
                else:
                    tmp = r
                    tmp[0] = list(pSet[r[1]])[0]
                    ws.writerow(tmp)
```

球員位置重整

```python
#clean Pos with C, F and G
for y in range(2011,2018):    
    nba = pd.read_csv('./NF_'+str(y)+'-'+str(y+1)+'_pid.csv')
    #print('./NF_'+str(y)+'-'+str(y+1)+'_pid.csv')
    nba['Pos'] = nba['Pos'].replace(to_replace=r'^.+F', value='F', regex=True)
    nba['Pos'] = nba['Pos'].replace(to_replace=r'^.+G', value='G', regex=True)
    print(nba)
    nba.to_csv('./NF_'+str(y)+'-'+str(y+1)+'_pid.csv', index = 0)
```

Only preserve the total data if a player stay more than one team

```python
#only preserve the total data if a player stay more than one team
index_number = [625, 551, 573, 611, 651, 578, 595, 664]
for y in range(2010,2018): #2011-2018
    nba = pd.read_csv('./NF_'+str(y)+'-'+str(y+1)+'_pid.csv')
    print('./NF_'+str(y)+'-'+str(y+1)+'_pid.csv')
    #nba = nba.set_index("pID")
    repeat_id = 0;
    for i in range(1, index_number[y-2010]):
        #print(str(i)+' '+str(nba.loc[i].pID))
        if nba.loc[i].pID == nba.loc[repeat_id].pID :
            #print (str(nba.loc[i].pID)+' '+nba.loc[i].Name+' '+nba.loc[i].Team)
            if nba.loc[i].Team != "TOT":
                nba = nba.drop(i)
            elif nba.loc[repeat_id].Team != "TOT":
                nba = nba.drop(repeat_id)
                repeat_id = i
        else:
            repeat_id = i
    #print(nba)
    nba.to_csv('./NF_'+str(y)+'-'+str(y+1)+'_pid.csv', index = 0)
```

增加進階feature

```python
#add features
for y in range(2011,2018):    
    nba = pd.read_csv('./NF_'+str(y)+'-'+str(y+1)+'_pid.csv')
    
    
    nba['EFF'] = ( nba['PTS'] + nba['TRB'] + nba['AST']  + nba['STL'] + nba['BLK'] - nba['FG'] - nba['FT'] - nba['TOV'] ) / nba['Game']
    
    nba['GmSc'] = nba['PTS'] + 0.4*nba['FG'] - 0.7*nba['FGA'] - 0.4*(nba['FTA']- nba['FT']) + 0.7*nba['ORB'] + 0.3*nba['DRB']+ nba['STL'] + 0.7*nba['AST'] + 0.7*nba['BLK'] - 0.4*nba['PF'] - nba['TOV']

    print(nba)
    
    nba.to_csv('./removeColumn_'+str(y)+'-'+str(y+1)+'_pid.csv', index = 0)

```
