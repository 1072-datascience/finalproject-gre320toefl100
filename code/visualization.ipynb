```python
import pandas as pd
data = pd.read_csv("../data/all_data.csv")
data.head()
```

```python
data.info()
```

```python
data_without_pid = data.drop(['pID','Name','Pos','Team'], axis=1)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
data_without_pid.hist(bins=20, figsize=(20,15))
plt.show()
```

```python
test = data[(data[['Salary']] != 0).all(axis=1)]
test2 = test[(test[['Pos']] != 'F-C').all(axis=1)]
```

```python
sns.set(font_scale=1)
sns.catplot(x="Pos", y="Salary", kind="swarm", data=test2)
```

```python
sns.boxplot(x="Pos", y="Salary", data=test2)
```

```python
data_without_pid.cov()
```

```python
import numpy as np
corr = test.corr()
f, ax = plt.subplots(figsize=(100, 60))

mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True


sns.set(font_scale=2.8)
sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="coolwarm",fmt='.3f', mask=mask,
                 linewidths=.05)
```

```python
# Scatter Plot
sns.set(font_scale=1)
plt.scatter(data['last_year_salary'], data['Salary'],
            alpha=0.4, edgecolors='c')

plt.xlabel('last_year_salary')
plt.ylabel('Salary')
#plt.title('Wine Sulphates - Alcohol Content',y=1.05)


# Joint Plot
jp = sns.jointplot(x='last_year_salary', y='Salary', data=data,
                   kind='reg', space=0, size=5, ratio=4)
```
